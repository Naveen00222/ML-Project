```python
# importing the necessary libraries..
import pandas as pd
```

```python
# reading the dataset
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\ML Project\datasets\diabetes_dataset.csv")
df
```

```python
# sampling the dataset to 5000 samples
data = df.sample(n=5000,random_state=42)
data
```

```python
# checking for object datatypes...
data.info()
```

```python
# checking for null values if any...
data.isna().sum()
```

```python
# checking for duplicates...
data.duplicated().sum()
```

```python
# removing unwanted columns...
data = data.drop(['ethnicity','education_level','income_level','employment_status','diabetes_stage','diagnosed_diabetes'],axis=1)
data.info()
```

```python
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting figure size
plt.figure(figsize=(18, 10))

# Age vs Diabetes Risk
plt.subplot(2, 3, 1)
sns.scatterplot(x='age', y='diabetes_risk_score', hue='gender', data=data)
plt.title('Age vs Diabetes Risk (by Gender)')

# BMI vs HbA1c
plt.subplot(2, 3, 2)
sns.scatterplot(x='bmi', y='hba1c', hue='smoking_status', data=data)
plt.title('BMI vs HbA1c (by Smoking Status)')

# Distribution of Fasting Glucose
plt.subplot(2, 3, 3)
sns.histplot(data['glucose_fasting'], bins=30, kde=True, color='teal')
plt.title('Fasting Glucose Distribution')

# Boxplot of BMI by Gender
plt.subplot(2, 3, 4)
sns.boxplot(x='gender', y='bmi', data=data)
plt.title('BMI by Gender')

# Correlation Heatmap of Key Health Metrics
plt.subplot(2, 3, (5, 6))
corr_cols = [
    'age', 'bmi', 'glucose_fasting', 'hba1c',
    'insulin_level', 'cholesterol_total', 'diabetes_risk_score'
]
sns.heatmap(data[corr_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')

# Adjust layout to avoid overlaps
plt.tight_layout()
plt.show()
```

```python
# Encoding......
```

```python
data['gender'].unique()
```

```python
data['smoking_status'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc1=LabelEncoder()
lab_enc2=LabelEncoder()
data['gender']=lab_enc1.fit_transform(data['gender'])
data['smoking_status']=lab_enc2.fit_transform(data['smoking_status'])
data
```

```python
# using train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['diabetes_risk_score'],axis=1)
y=data['diabetes_risk_score']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# creating the models
```

```python
# importing necessary model packages and modules....
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

```python
# random forest model...
rf_mod = RandomForestRegressor()
rf_mod.fit(x_train,y_train)
print("Test score  : ", rf_mod.score(x_test,y_test))
print("Train score : ", rf_mod.score(x_train,y_train))
```

```python
# DecisionTree Model...
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train,y_train)
print("Test score  : ", dt_model.score(x_test,y_test))
print("Train score : ", dt_model.score(x_train,y_train))
```

```python
# KNeighbors Model...
knn_model = KNeighborsRegressor(n_neighbors=9)
knn_model.fit(x_train,y_train)
print("Test score  : ", knn_model.score(x_test,y_test))
print("Train score : ", knn_model.score(x_train,y_train))
```

```python
# Adaboost Model...
ad_model = AdaBoostRegressor()
ad_model.fit(x_train,y_train)
print("Test score  : ", ad_model.score(x_test,y_test))
print("Train score : ", ad_model.score(x_train,y_train))
```

```python
# xgboost model...
xg_model=XGBRegressor()
xg_model.fit(x_train,y_train)
print("Test score  : ", xg_model.score(x_test,y_test))
print("Train score : ", xg_model.score(x_train,y_train))
```

```python
# Hypertuning to get best parameters...
from sklearn.model_selection import GridSearchCV
gscv=GridSearchCV(xg_model,{'n_estimators':[100,200,300,400,500],'learning_rate':[0.05,0.10,0.20]},
                cv=5,return_train_score=True)
gscv.fit(x_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_model = gscv.best_estimator_
best_model
```

```python
import pickle

# Save model, scaler, and encoders
pickle.dump(xg_model, open('diabetes_model.pkl', 'wb'))
pickle.dump(scaler, open('scaler.pkl', 'wb'))
pickle.dump(lab_enc1, open('label_enc_gender.pkl', 'wb'))
pickle.dump(lab_enc2, open('label_enc_smoking.pkl', 'wb'))
```
