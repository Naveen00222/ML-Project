```python
import pandas as pd
df=pd.read_csv('crop_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
top_states=df.groupby('State_Name')['Production'].sum().sort_values(ascending=False).head(10)
top_states.plot(kind='bar', title='Top 10 States by Total Production', figsize=(10, 5), color='orange')
plt.ylabel('Total Production')
plt.show()
```

```python
top_crops=df.groupby('Crop')['Production'].sum().sort_values(ascending=False).head(10)
top_crops.plot(kind='bar',title='Top 10 Crops by Production',figsize=(10, 5),color='green')
plt.ylabel('Total Production')
plt.show()
```

```python
df=pd.get_dummies(df,columns=['State_Name','District_Name','Season','Crop'],drop_first=True)
```

```python
x=df.drop('Production', axis=1)
y=df['Production']
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
df=df.astype(int)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
lr_model=LinearRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test,y_test)
```

```python
dt_model=DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
dt_model.score(X_test,y_test)
```

```python
rf_model=RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
rf_model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
lr_model=LinearRegression()
lr_scores=cross_val_score(lr_model,x,y,cv=5,scoring='r2')
print("Linear Regression CV Score:",np.mean(lr_scores))
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_scores = cross_val_score(dt_model,x,y,cv=5,scoring='r2')
print("Decision Tree CV Score:",np.mean(dt_scores))
```

```python
rf_model=RandomForestRegressor(random_state=42)
rf_scores=cross_val_score(rf_model,x,y,cv=5,scoring='r2')
print("Random Forest CV Score:",np.mean(rf_scores))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
model=RandomForestRegressor()
params={
    'n_estimators': [50, 100],
    'max_depth': [10, None]
}
grid=GridSearchCV(model, params, cv=5, scoring='r2')
grid.fit(X_train, y_train)

print("Best Params:",grid.best_params_)
print("Best Score:",grid.best_score_)
```

```python
final_model=RandomForestRegressor(max_depth=10,n_estimators=50,random_state=42)
final_model.fit(X_train, y_train)
```
